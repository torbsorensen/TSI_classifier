```python
import xarray as xr
import pandas as pd
import datetime
import numpy as np
```

```python
data = pd.read_csv('predictions/prediction_CSV/20140723.csv',header=None)
```

```python
data
```

```python
times = []
prediction = []
predconf = []
fogconf = []
iceconf = []
noconf = []

lon = 20.592342
lat = 84.235134

for i in range(len(data)):
    times.append(pd.to_datetime(str(data.iloc[i,0])[:-2]))
    prediction.append(int(data.iloc[i,1]))
    predconf.append(data.iloc[i,2])
    fogconf.append(data.iloc[i,3])
    iceconf.append(data.iloc[i,4])
    noconf.append(data.iloc[i,5])
```

```python
ds = xr.Dataset()
```

```python
i=0

for i in range(5):
    new_ds = xr.Dataset(
         {
             "prediction": (["time"], prediction),
             "prediction_confidence": (["time"], predconf),
             "fog_confidence": (["time"], fogconf),
             "ice_confidence": (["time"], iceconf),
             "no_confidence": (["time"], noconf)
         },
         coords={
             "time": times,
         },
     )
    
    xr.concat(ds, new_ds)
    
    i = i+1

```
